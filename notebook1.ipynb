```python
# 1️⃣  Load Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
# Load Data
gscpi = pd.read_excel("gscpi_data.xls", sheet_name="Data", parse_dates=True)
wb_stress = pd.read_excel("Stress index tables.xlsx", sheet_name="global", parse_dates=True)

# Inspect
print(gscpi.head())
print(wb_stress.head())
```

```python
gscpi.Date = pd.to_datetime(gscpi.Date)
```

```python
wb_stress.Departure_YearMonth = pd.to_datetime(wb_stress.Departure_YearMonth, format="%Y%m%d")
```

```python
gscpi.columns = ["Date", "GSCPI"]
wb_stress.columns = ["Date", "WB_Stress"]
wb_stress = wb_stress.set_index("Date")
data_nearest = pd.merge_asof(
    gscpi.sort_values("Date"), 
    wb_stress.sort_index(), 
    on="Date", 
    direction="nearest", 
    tolerance=pd.Timedelta("30D")
)
data_nearest.head()
```

```python
# Clean & Align
gscpi.columns = ["Date", "GSCPI"]
wb_stress.columns = ["Date", "WB_Stress"]
data = pd.merge(gscpi, wb_stress, on="Date", how="inner").sort_values("Date")
```

```python
data.head()
```

```python

# 5️⃣  Visualize
sns.lineplot(data=data, x="Date", y="GSCPI", label="NY Fed GSCPI")
sns.lineplot(data=data, x="Date", y="WB_Stress", label="World Bank Stress")
plt.title("Global Supply Chain Stress Indices Over Time")
plt.show()

# 6️⃣  Summary Stats
print(data.describe())
data.corr()

# 7️⃣  (Later) Merge Insurance / Loss Data
# claims = pd.read_csv("LossRatios.csv")
# data = pd.merge(data, claims, on="Date", how="inner")

# 8️⃣  R
```
